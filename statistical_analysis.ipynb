```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Crop': ['Maize', 'Rice', 'Beans'],
    'Yield': [3.5, 4.2, 2.8]
})

df.plot(kind='bar', x='Crop', y='Yield', legend=False)
plt.title("Crop Yield")
plt.ylabel("Tonnes per Hectare")
plt.show()
```

```python
import pandas as pd
import mindspore
from mindspore import Tensor, context
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set MindSpore context to CPU mode
context.set_context(mode=context.GRAPH_MODE, device_target="CPU")

# Load the CSV file
df = pd.read_csv('/content/sample_data/data_soil_crop.csv')
# Load CSV data
def load_data(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError("CSV file not found!")
    df = pd.read_csv('/content/sample_data/data_soil_crop.csv')
    print("Data Loaded Successfully. Columns are:\n", df.columns.tolist())
    return df
# Basic statistical summary
print("Summary Statistics:\n", df.describe())

# Group by soil type
#print("\nAverage by Soil Type:\n", df.groupby('Soil Type').mean())
numeric_cols = df.select_dtypes(include='number').columns
grouped = df.groupby('Crop Type')[numeric_cols].mean()
print(grouped)

# Correlation matrix
print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))

# Save plots in 'plots' directory
os.makedirs("plots", exist_ok=True)

# Histogram
df.hist(bins=10, figsize=(10, 5))
plt.suptitle("Histogram of Soil Features")
plt.savefig("plots/histograms.png")

# Boxplot by soil type
plt.figure(figsize=(6, 4))
sns.boxplot(x='Soil Type', y='Crop Type', data=df)
plt.title("Boxplot of Crop Type by Soil Type")
plt.savefig("plots/crop_type_by_soil_type.png")

# Scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Crop Type', y='Moisture', hue='Soil Type', data=df)
plt.title("Crop Type vs Moisture")
plt.savefig("plots/Crop Type_vs_Moisture.png")

# Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("plots/correlation_heatmap.png")

print("\nüìÅ All plots saved to the 'plots/' folder.")
```

```python
import pandas as pd
import mindspore
from mindspore import Tensor, context
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set MindSpore context to CPU mode
context.set_context(mode=context.GRAPH_MODE, device_target="CPU")

# Load the CSV file
df = pd.read_csv('/content/sample_data/agriculture_dataset.csv')
# Load CSV data
def load_data(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError("CSV file not found!")
    df = pd.read_csv('/content/sample_data/agriculture_dataset.csv')
    print("Data Loaded Successfully. Columns are:\n", df.columns.tolist())
    return df
# Basic statistical summary
print("Summary Statistics:\n", df.describe())

# Group by crop type
#print("\nAverage by Crop_Type:\n", df.groupby('Crop_Type').mean())
numeric_cols = df.select_dtypes(include='number').columns
grouped = df.groupby('Crop_Type')[numeric_cols].mean()
print(grouped)

# Correlation matrix
print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))

# Save plots in 'plots' directory
os.makedirs("plots", exist_ok=True)

# Histogram
df.hist(bins=10, figsize=(10, 5))
plt.suptitle("Histogram of Crop Features")
plt.savefig("plots/histograms.png")

# Boxplot by crop type
plt.figure(figsize=(6, 4))
sns.boxplot(x='Crop_Type', y='Expected_Yield', data=df)
plt.title("Boxplot of Crop_Type by Expected_Yield")
plt.savefig("plots/Crop_Type_by_Expected_Yield.png")

# Scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Crop_Type', y='Expected_Yield', hue='Rainfall', data=df)
plt.title("Crop_Type vs Expected_Yield")
plt.savefig("plots/Crop_Type_vs_Expected_Yield.png")

# Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("plots/correlation_heatmap.png")

print("\nüìÅ All plots saved to the 'plots/' folder.")
```
